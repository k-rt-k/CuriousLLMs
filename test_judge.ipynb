```python
import numpy as np
import torch
from llm_as_a_judge import GeminiJudge

from dotenv import load_dotenv
load_dotenv()

judge = GeminiJudge()
```

```python
Questions = [
    "What is the maximum value of the function f(x) = -2x^2 + 4x + 1 on the interval [0, 3]?",
    "If a triangle has sides of length 3, 4, and 5, what is its area?",
    "Solve for x: 2x + 3 = 7"
]

References = [
    "The maximum value is 3.",
    " 6.",
    " x = 2."
]

Answers = [
    "Note that the function is a downward-opening parabola. We can find the argmax by taking the derivative and setting it to zero: f'(x) = -4x + 4 = 0 => x = 1. Plugging back into f(x), we get f(1) = 3. Therefore, the maximum value is 3.",
    "The triangle with sides 3, 4, and 5 is a right triangle. Thus, we can use the formula for the area of a right triangle: A = (1/2) * base * height = (1/2) * 3 * 4 = 6.",
    "Rearranging the equation gives us 2x = 4, so x = 3."
]
```

```python
judge.judge_single(question=Questions[0], answer=Answers[1], reference=References[0])
```

```python
judge.judge_synchronous(questions=Questions, answers=Answers, references=References)
```

```python
judge.judge_batch(questions=Questions, answers=Answers, references=References)
```
